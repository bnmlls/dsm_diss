```python
from medcat.ner import transformers_ner
from medcat.config_transformers_ner import ConfigTransformersNER, General
from medcat.cat import CAT
from medcat.cdb import CDB
from medcat.vocab import Vocab
import os
import logging
from medcat.preprocessing.tokenizers import TokenizerWrapperBPE, TokenizerWrapperBERT
from tokenizers import ByteLevelBPETokenizer
```

```python
# Specify the directory containing the model data
DATA_DIR = '/Users/ben/unsupervised_trained_model_pack/medcat_model_pack_f4d4528c96b2b0f7/'

# Define paths for the CDB and Vocabulary
vocab_path = os.path.join(DATA_DIR, "vocab.dat")
cdb_path = os.path.join(DATA_DIR, "cdb.dat")
```

```python
cdb = CDB.load(cdb_path)
```

```python
from medcat.config_transformers_ner import ConfigTransformersNER, General


# Specify the transformer model name
transformer_model = "emilyalsentzer/Bio_ClinicalBERT"

# Create a transformer-based configuration
transformer_config = ConfigTransformersNER(
    general=General(
        name="medcat_transformer_model",
        model_name=transformer_model,
        description=f"Transformer-based NER using MedCAT with {transformer_model}",
        pipe_batch_size_in_chars=20000000,
        ner_aggregation_strategy="simple",
        test_size=0.2,
        last_train_on=None,
        verbose_metrics=False,
        log_level= None#empty
    ),
    
    spacy_model='en_core_web_md'  
)
```

```python
import os
from medcat.ner.transformers_ner import TransformersNER

# Set the environment variable to flatten parameters
os.environ["MLFLOW_FLATTEN_PARAMS"] = "true"

# Now create an instance of TransformersNER and train the model
transformers_ner = TransformersNER(cdb=cdb, config=transformer_config)
transformers_ner.train(json_path="/Users/ben/Documents/mscdatascience/Dissertation/MedCAT_Export.json")
```
